Uploader le fichier products.csv. Rep√©rer les valeurs aberrantes dans le fichier (ordre de grandeur : plusieurs centaines).

```
from pyspark.sql.functions import col
```

```
df = spark.read.csv("/FileStore/tables/SPN_products.csv", header=True, inferSchema=True)
```

```
df_negative = df.filter(
    (col("fat_100g") < 0) |
    (col("saturated-fat_100g") < 0) |
    (col("sugars_100g") < 0) |
    (col("fiber_100g") < 0) |
    (col("proteins_100g") < 0) |
    (col("salt_100g") < 0) |
    (col("sodium_100g") < 0) |
    (col("autre") < 0)
)

df_negative.show()
print(df_negative.count())
```

```
df_sum_more_than_100 = df.filter(
    (
        col("fat_100g") +
        col("sugars_100g") +
        col("fiber_100g") +
        col("proteins_100g") +
        col("salt_100g") +
        col("autre")
    ) > 100
)

df_sum_more_than_100.show()
print(df_sum_more_than_100.count())
```

```
df = df.fillna(0)
```

```
fat_lt_saturated_df = df.filter(
    col("fat_100g") < col("saturated-fat_100g")
)

print(fat_lt_saturated_df.count())
```
